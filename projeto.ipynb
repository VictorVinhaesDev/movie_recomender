### Vendo colunas em JSON

```python
# Função para verificar se uma string é JSON válida
import json

def is_json(text):
    try:
        json.loads(text)
        return True
    except (ValueError, TypeError):
        return False

# Função para verificar se uma coluna contém JSON válido
def check_columns_for_json(df):
    for column in df.columns:
        # Verifica se pelo menos um valor na coluna é JSON válido
        is_column_json = df[column].apply(is_json).any()
        print(f"{column}: {is_column_json}")

```

```python
# Chame a função para verificar as colunas do DataFrame
check_columns_for_json(base)
```

### Filtrando a coluna genres e keywords
- Vamos pegar apenas o "name" das colunas
- Como a coluna "genre" e "keywords" estão em formato JSON temos que tratar ela 

```python
import json
def convert(text):
    lista = []
    for item in json.loads(text):
        lista.append(item['name'])
    return lista
```

#### Para a coluna "cast" 
- Quero apenas os 3 primeiros nomes
- Vou ter que alterar a função

```python
import json
def convert(text):
    lista = []
    counter = 0
    for item in json.loads(text):
        if(counter <= 3):
            lista.append(item['name'])
        else:
            break
    return lista
```

### Para a coluna "crew" queremos pegar o nome do diretor para colocar em outra coluna
- Filtraremos aonde o trabalho seja diretor e pegaremos o nome
- Desta vez criaremos uma nova coluna e não substituiremos

```python
import json
def convert(text):
    lista = []
    counter = 0
    for item in json.loads(text):
        if(item["job"] == "Director"):
            lista.append(item['name'])
    return lista
```

### Vamos usar a lib "nltk" tecnicas de linguagem natural (NLP
- https://www.nltk.org/
- Serve para simplificar as palavras parecidas para melhorar o machine learn
- Existem várias mas aqui vamos utilizar o "PorterStemmer" desta lib é uma ferramenta popular de stemming que reduz palavras à sua raiz básica ou "stem". Isso é útil em muitas tarefas de processamento de linguagem natural, como análise de texto e recuperação de informações, onde a variação das palavras deve ser tratada de forma unificada.

```python
import nltk
from nltk.stem import PorterStemmer
# Inicializar o stemmer
stemmer = PorterStemmer()
```

```python
# Exemplo do que o steammer faz, todas as palavras viraram "love"
arr = ["loved", "lovely", "loving", "love"]
for i in arr:
    print(stemmer.stem(i))
```

```python
# Função para usar no database
def fun_stemmer(text):
    list = []
    for i in text.split(): # precisa do split para não pegar cada letra da palavra
        list.append(stemmer.stem(i))
    return " ".join(list) # join serve para retornar a frase junta e não separadas em um array
```

```python
base['tags'] = base['tags'].apply(fun_stemmer)
```

### Vectorization
- A transformação de dados em vetores numéricos para que modelos possam ser treinados ou fazer previsões. Um exemplo comum é a vetorização de texto usando "CountVectorizer" do scikit-learn.
- https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html#sklearn.feature_extraction.text.CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(
                            stop_words = "english", # Padrão "none". Remove palavras comuns que geralmente não são informativas.
                            max_features = 5000 # Padrão "none".Fica em 5mil colunas no maximo a matriz, Mantém apenas os max_features termos mais frequentes.
                            )
```

```python
# Ajustando o vetor de contagem aos textos
texts = base["tags"]
X_vec = vectorizer.fit_transform(texts)
```

```python
X_vec.shape
```

```python
# Mostrando os recursos (features) extraídos
print(vectorizer.get_feature_names_out())
```

```python
# 5 mil igual definimos no "max_features"
len(vectorizer.get_feature_names_out())
```

```python
# Mostrando a matriz de contagem
print(X_vec.toarray())
```

```python
# IMPORTANTE, passando para o array para ficar mais fácil de usar
X_vec = X_vec.toarray()
```

### Usaremos o "cosine_similarity" para recomendar os filmes semelhantes
- https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.cosine_similarity.html#sklearn.metrics.pairwise.cosine_similarity
- A similaridade do cosseno (cosine similarity) é uma medida usada para determinar o quão similares são dois vetores, independentemente de sua magnitude. No contexto de aprendizado de máquina e processamento de linguagem natural (NLP), essa métrica é frequentemente utilizada para comparar documentos, frases ou qualquer tipo de texto representado como vetores.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Calcula as similaridades de todas as linhas entre elas mesmas da coluna que escolhemos, no caso a "tags"
similarity = cosine_similarity(X_vec)
```

```python
# Isso que vamos usar para comparar com os próprios filmes
similarity
```

### Explicando a função abaixo

```python
base[base["title"] == "Batman"] # por isso pegamos o indice
```

```python
base[base["title"] == "Batman"].index[0]
```

```python
def recommend(movie):
    movie_index = base[base["title"] == movie].index[0] # temos que pegar o valor do primeiro indice valor(index[0]) pois tem filmes que tem titulos iguais dai ia gerar recomendações do mesmo filme só que em outras linguas talvez
    similarity_movie = similarity[movie_index] # pegamos a similaridade do filme escolhido pelo indice, ele calcula a similaridade dele a respeito de toda base que já escolhemos antes pelo coseno -> "cosine_similarity(X_vec)"
    
    
    movies_list_index = sorted(list(enumerate(similarity_movie)),reverse = True, key = lambda x: x[1])[1:6] # vai puxar a similaridade de toda a lista´, pegamos do indice 1 até 6 pois o 0 é ele mesmo então não faria sentido.
    # o "reverse = true" do sorted vai pegar da ordem do maior para o menor
    # e a "key = lambda x: x[1]" serve para informar que queremos o segundo valor para escolher como ordernar no caso o primeiro valor é o indice "0" o segundo é a similaridade que é o "1"

    
    for i in movies_list_index:  # printando os valores
        print(base.iloc[i[0]].title) # passando "i[0]" pois assim vem só o indice do filme sem a % de similaridade
        #  A função "iloc" retorna a linha baseada no número do indice
```

```python
recommend("Avatar")
```

```python
recommend("Batman")
```
